```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # Choose any style you like
```

```python

labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
 
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs,labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
plt.show()
```

```python
explode = [0.1, 0.1, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors,shadow=True,startangle=180)
plt.title("Exploded Pie Example")
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=140
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode
)
plt.title("Career Run Share")
plt.show()
```
